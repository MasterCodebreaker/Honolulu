```python
import pandas as pd
import plotly.graph_objects as go
import numpy as np
import tadasets

output = "point_color"
start_end_ = "vectors"
```

```python
ok = output + ".csv"
df = pd.read_csv(ok)
ok = start_end_ + ".csv"
start_end_1 = pd.read_csv(ok)

#1111111111111
L = [1,26,67,93] # WRITE INPUT HERE

x = df.iloc[:,0]
y = df.iloc[:,1]
z = df.iloc[:,2]
c_1 = df.iloc[:,3]
co_1 = np.mod(c_1,1.0)
fig = go.Figure(data=[go.Scatter3d(
    x=x,
    y=y,
    z=z,
    mode='markers',
    text = [str(i) for i in co_1],
    marker=dict(
        size=4,
        color = np.mod(co_1,1.0),             # set color to an array/list of desired values
        colorscale='HSV',
        # choose a colorscale
        opacity=0.6
    )
)])

fig.add_trace(
    go.Scatter3d(
    x = df.iloc[L[:2]].iloc[:,0],
    y = df.iloc[L[:2]].iloc[:,1],
    z = df.iloc[L[:2]].iloc[:,2], mode = "markers",
      
      marker=dict(
        #c = "black",
        size=10,
        color = 0,             # set color to an array/list of desired values
        colorscale='magma',
        # choose a colorscale
        opacity=0.6
    )))


fig.add_trace(
    go.Scatter3d(
    x = df.iloc[L[2:]].iloc[:,0],
    y = df.iloc[L[2:]].iloc[:,1],
    z = df.iloc[L[2:]].iloc[:,2], mode = "markers",
      
      marker=dict(
        #color = "black",
        size=10,
        color = 1,             # set color to an array/list of desired values
        colorscale='magma',
        # choose a colorscale
        opacity=0.6
    )))


fig.show()
```
